### From the above output, we can see above most have the columns have missing vales of more than 50%Â¶

### To find the exact number,Now we will check the percentage of nan values present in each feature

```python
# list of features which has missing values
features_with_na=[features for features in Terrorism.columns if Terrorism[features].isnull().sum()>1]

# print the feature name and the percentage of missing values
for feature in features_with_na:
    print(feature, np.round(Terrorism[feature].isnull().mean(), 4),  ' % missing values')
```
